```python
# Dependencies and Setup
import pandas as pd

# File to Load
school_data_to_load = "../Resources/schools_complete.csv"
student_data_to_load = "../Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# Display the Data Set
df = school_data_complete
df
```

# District Summary
* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math and reading (% Overall Passing)

* Create a dataframe to hold the above results
* Optional: give the displayed data cleaner formatting

```python
# 1. Calulate the Total Number of Schools
## Step 1: Create an array of unique school names
unique_school_names = school_data_complete["school_name"].unique()
unique_school_names

## Step 2: Count the number of unique schools
total_schools = len(unique_school_names)
print(total_schools)
```

```python
# 2. Caluclate the total number of students
total_students = school_data_complete['student_name'].count()
total_students
```

```python
# 3. Calculate the total budget
total_budget = sum(school_data_complete['budget'].unique())
total_budget
```

```python
# 4. Calculate the average math score
average_math = school_data_complete['math_score'].mean()
average_math
```

```python
# 4. Calculate the average reading score
average_reading = school_data_complete['reading_score'].mean()
average_reading
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
per_pass_math = school_data_complete.loc[school_data_complete['math_score'] >= 70]['math_score'].count()/total_students
print(per_pass_math)
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
per_pass_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70]['reading_score'].count()/total_students
print(per_pass_reading)
```

```python
# Calculate the percentage of students who passed math and reading (% Overall Passing)
per_overall_pass = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)]['student_name'].count()/total_students
print(per_overall_pass)
```

```python
# Create a dataframe to hold the above results
district_summary = pd.DataFrame({
    
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math],
    "Average Reading Score": [average_reading],
    "% Passing Math":[per_pass_math],
    "% Passing Reading": [per_pass_reading],
    "% Overall Passing": [per_overall_pass]
    })
district_summary.head()
```

```python
# Give the displayed data cleaner formatting
district_summary.style.format({"Total Students": "{:,}",
                       "Total Budget": "${:,.2f}", 
                       "Average Math Score": "{:.1f}", 
                       "Average Reading Score": "{:.1f}", 
                       "% Passing Math": "{:.1%}", 
                       "% Passing Reading": "{:.1%}", 
                       "% Overall Passing": "{:.1%}"})
```

```python
# Group the data frame by school name 
grouped_schools = school_data_complete.groupby(['school_name'])
grouped_schools
```

```python
# Display the school type for each school
school_type = grouped_schools['type'].first()
school_type
```

```python
# Display the total number of students for each school
total_students_per_school = grouped_schools.size()
print(total_students_per_school)
```

```python
# Display the budget for each school
budget_per_school = grouped_schools['budget'].first()
print(budget_per_school)
```

```python
# Display the budget per student for each school
budget_per_student = budget_per_school/total_students_per_school
print(budget_per_student)
```

```python
# Display the average math score for each school
avg_math_by_school = grouped_schools['math_score'].mean()
print(avg_math_by_school)
```

```python
# Display the average reading score for each school
avg_reading_by_school = grouped_schools['reading_score'].mean()
print(avg_reading_by_school)
```

```python
# Calculate the percentage of students with a passing math score (70 or greater) for each school
pass_math_by_school = school_data_complete[school_data_complete['math_score']>=70].groupby(['school_name']).size()
per_pass_math_by_school = (pass_math_by_school/total_students_per_school)
print(per_pass_math_by_school)
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater) for each school
pass_reading_by_school = school_data_complete[school_data_complete['reading_score']>=70].groupby(['school_name']).size()
per_pass_reading_by_school = (pass_reading_by_school/total_students_per_school)
print(per_pass_reading_by_school)
```

```python
# Calculate the percentage of students who passed math and reading (% Overall Passing) for each school
overall_pass_by_school = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)].groupby(['school_name']).size()
per_overall_pass_by_school = overall_pass_by_school/total_students_per_school
print(per_overall_pass_by_school)
```

```python
# Create a dataframe to hold the above results
school_summary = pd.DataFrame({  
    "School Type": school_type,
    "Total Students": total_students_per_school,
    "Total Budget": budget_per_school,
    "Budget Per Student": budget_per_student, 
    "Average Math Score": avg_math_by_school,
    "Average Reading Score": avg_reading_by_school,
    "% Passing Math":per_pass_math_by_school,
    "% Passing Reading": per_pass_reading_by_school,
    "% Overall Passing": per_overall_pass_by_school
    })
```

```python
# Reformat the school summary
school_summary['Total Students'] = school_summary['Total Students'].map("{:,}".format)
school_summary['Total Budget'] = school_summary['Total Budget'].map("${:,.2f}".format)
school_summary['Budget Per Student'] = school_summary['Budget Per Student'].map("${:,.2f}".format)
school_summary['Average Math Score'] = school_summary['Average Math Score'].map("{:,.1f}".format)
school_summary['Average Reading Score'] = school_summary['Average Reading Score'].map("{:,.1f}".format)
school_summary['% Passing Math'] = school_summary['% Passing Math'].map("{:,.2%}".format)
school_summary['% Passing Reading'] = school_summary['% Passing Reading'].map("{:,.2%}".format)
school_summary['% Overall Passing'] = school_summary['% Overall Passing'].map("{:,.2%}".format)

school_summary
```

# Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math (The percentage of students that passed math.)
  * % Passing Reading (The percentage of students that passed reading.)
  * % Overall Passing (The percentage of students that passed math **and** reading.)

```python
# Create spending bins
bins = [0, 584.999, 614.999, 644.999, 999999]
group_name = ["< $585", "$585 - 614", "$615 - 644", "> $644"]
school_data_complete['spending_bins'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], bins, labels = group_name)

# Group by spending
by_spending = school_data_complete.groupby('spending_bins')

# Calculations
avg_math = by_spending['math_score'].mean()
avg_read = by_spending['reading_score'].mean()
pass_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()
pass_read = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()
overall = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()

            
# Build the dataframe            
scores_by_spend = pd.DataFrame({
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "% Overall Passing": overall
})

scores_by_spend.index.name = "Per Student Budget"
scores_by_spend = scores_by_spend.reindex(group_name)

# Reformat
scores_by_spend.style.format({'Average Math Score': '{:.1f}', 
                              'Average Reading Score': '{:.1f}', 
                              '% Passing Math': '{:.1%}', 
                              '% Passing Reading':'{:.1%}', 
                              '% Overall Passing': '{:.1%}'})
```
